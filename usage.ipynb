```python
from python_ace import atoms_to_ace
import numpy as np
import ase
```

```python
atoms = ase.io.read('data/example.extxyz')
ace = atoms_to_ace(atoms)
ace.shape
```

```python
#!pip install scikit-learn
#!pip install matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
ace_pca = pca.fit_transform(ace[0])

x,y = ace_pca.T
plt.scatter(x, y, s=1, c='green')
plt.title('PCA of ACE descriptor')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
